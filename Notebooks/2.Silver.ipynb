## **Creating a parameterized notebook**

###  **Creating Parameters**

```sql
%python
dbutils.widgets.text("sourcefolder","netflix_directors")
dbutils.widgets.text("targetfolder","netflix_directors")
```

### Creating Variables

```sql
%python
src_folder = dbutils.widgets.get("sourcefolder")
tar_folder = dbutils.widgets.get("targetfolder")
```

```sql
%python
print(src_folder)
print(tar_folder)
```

```sql
%python
file_path = f"abfss://bronze@netflixdatastorage.dfs.core.windows.net/{src_folder}"
```

```sql
%python
df = spark.read.format("csv")\
          .option("header",True)\
          .option("inferSchema",True)\
          .load(file_path)
```

```sql
%python
display(df)
```

```sql
%python
f_path = f"abfss://silver@netflixdatastorage.dfs.core.windows.net/{tar_folder}"
```

```sql
%python
df.write.format("delta")\
    .mode("append")\
    .option('path', f_path)\
    .save()
```
